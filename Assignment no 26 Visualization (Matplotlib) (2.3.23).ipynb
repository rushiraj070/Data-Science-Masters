```python
'''Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.'''

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

#Note: Also add title, xlabel, and ylabel to the plot.
```

```python
plt.scatter(x,y, c = "r")
plt.title("Scatter Plot X Vs Y")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
'''
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
#plt.figure()

plt.subplot(2,2,1)
plt.plot(x1,y1)

plt.subplot(2,2,2)
plt.plot(x2,y2)

plt.subplot(2,2,3)
plt.plot(x3,y3)

plt.subplot(2,2,4)
plt.plot(x4,y4)
```

```python
#plt.figure()

plt.subplot(2,2,1)
plt.plot(x1,y1, "ro")

plt.subplot(2,2,2)
plt.plot(x2,y2, "g--")

plt.subplot(2,2,3)
plt.plot(x3,y3, "kp")

plt.subplot(2,2,4)
plt.plot(x4,y4, "b*")
```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Ans - A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths 
    proportional to the values that they represent. The bars can be plotted vertically or horizontally.

    A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and 
    the other axis represents a measured value.
'''
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit, color = "y")
plt.title("Bar Graph for Company Vs Profit")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

```python
plt.barh(company,profit, color = "violet")
plt.title("Bar Graph for Profit Vs Company")
plt.ylabel("Company")
plt.xlabel("Profit")
plt.show()
```

```python
'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Ans - A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, 
    median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. 
    
    A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.
'''
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.figure()
plt.boxplot(box1)
plt.title("Box Plot 1")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.figure()
plt.boxplot(box2)
plt.title("Box Plot 2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

```python
plt.figure()

plt.subplot(3,2,1)
plt.boxplot(box1)
plt.title("Box Plot 1")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.subplot(3,2,2)
plt.boxplot(box2)
plt.title("Box Plot 2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```
